```python
import requests
import pandas as pd

# URL de ejemplo: Eventos culturales en JSON
url = "https://datos.madrid.es/egob/catalogo/206974-0-agenda-eventos-culturales-100.json"

# Hacer la petición GET
response = requests.get(url)

# Verificar si la respuesta es correcta (código 200)
if response.status_code == 200:
    data = response.json()  # Convertir a diccionario Python
    print(type(data))  # Mostrar el tipo de dato
    print(data)  # Mostrar los datos
else:
    print(f"Error en la petición: {response.status_code}")
```

```python
# Extraer la lista de eventos
eventos = data["@graph"]  # La información está dentro de "@graph"
```

```python
# Convertir a pandas
df_eventos = pd.DataFrame(eventos)

# Mostrar las primeras filas
df_eventos.head()

df_eventos.columns

# Eliminar columnas innecesarias
df_eventos = df_eventos[['id', 'title', 'description', 'free', 'price',
       'dtstart', 'dtend', 'time', 'excluded-days', 'uid',
       'link', 'event-location', 'address',
       'location', 'organization', 'audience']]

df_eventos


```

```python
# Obtener columnas a partir de la direccion (address)
df_eventos['latitude'] = df_eventos['location'].apply(lambda x: x['latitude'] if isinstance(x, dict) else None)
df_eventos['longitude'] = df_eventos['location'].apply(lambda x: x['longitude'] if isinstance(x, dict) else None)
df_eventos['postal-code'] = df_eventos['address'].apply(lambda x: x['area']['postal-code'] if isinstance(x, dict) and 'area' in x and 'postal-code' in x['area'] else None)
df_eventos['calle'] = df_eventos['address'].apply(lambda x: x['area']['street-address'] if isinstance(x, dict) and 'area' in x and 'street-address' in x['area'] else None)

# Display the updated dataframe
df_eventos.columns

df_eventos =  df_eventos[['id', 'title', 'description', 'free', 'price', 'dtstart', 'dtend',
       'time', 'excluded-days', 'uid', 'link', 'event-location', 'organization', 'audience', 'latitude', 'longitude', 'postal-code',
       'calle']]

df_eventos.head()
```

```python
remap_dict = {'nombre_evento': 'title', 'url_evento': 'link', 'codigo_postal': 'postal-code', 'direccion': 'calle', 'horario': 'time', 'organizacion': 'organization'}
```

```python
# Invert the dictionary: {old_name: new_name}
rename_dict_inverted = {v: k for k, v in remap_dict.items()}

# Rename columns using the inverted dictionary
df_eventos.rename(columns=rename_dict_inverted, inplace=True)

# Display result
print(df_eventos)

```

```python
df_eventos_2_db =  df_eventos[columns_to_keep]
df_eventos_2_db.head()
```
