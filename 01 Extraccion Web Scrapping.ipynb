```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
from datetime import date
import pandas as pd

# URL of the website
url = 'https://all.accor.com/ssr/app/ibis/hotels/madrid-spain/index.en.shtml?compositions=1&stayplus=false&snu=false&hideWDR=false&accessibleRooms=false&hideHotelDetails=false&dateIn=2025-03-01&nights=1&destination=madrid-spain'

# Set up the Selenium WebDriver
driver = webdriver.Chrome()  # Make sure you have the ChromeDriver installed and in your PATH
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Close the WebDriver
driver.quit()

# Find all hotel elements
hotels = soup.find_all('a', class_='title__link title__link')

# Extract hotel names, ratings, and prices
hotel_data = []
for hotel in hotels:
    name = hotel.get_text(strip=True)
    
    # Find the ratingscore
    rating_element = hotel.find_next('span', class_='ratings__score')
    rating = rating_element.get_text(strip=True).split('/')[0] if rating_element else 'N/A'
    
    # Find the price
    price_element = hotel.find_next('span', class_='booking-price__number mcp-price-number')
    price = price_element.get_text(strip=True) if price_element else 'N/A'
    
    hotel_data.append({'Name': name, 'Rating': rating, 'Price': price})

# Print the extracted data
for data in hotel_data:
    print(data)

# Save data into DataFrame
df = pd.DataFrame(hotel_data)
df['date'] = str(date.today())
df
```

```python
renama_dict = {'Name':'nombre_hotel', 'Rating':'estrellas', 'Price':'precio_noche', 'date':'fecha_reserva'}
# Rename columns using the inverted dictionary
df.rename(columns=renama_dict, inplace=True)

# Display result
print(df)

df.to_csv(r'datos_a_db\hoteles_competencia.csv', index=False)
```
